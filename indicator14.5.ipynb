```python
# Lag plot:

from pandas.plotting import lag_plot
lag_plot(con14_1['Value'])
plt.title("Lag plot of lag 1 for Continents")
plt.show()
```

```python
# Trend seasonal plot for Australia
from pylab import rcParams
import statsmodels.api as sm
plt.rcParams["figure.figsize"] = [14,6]
decomposition = sm.tsa.seasonal_decompose(con14_1_Aus['Value'], model='Additive', freq =2)
fig = decomposition.plot()
plt.title('Seasonal plot for Australia')
plt.show()
```
