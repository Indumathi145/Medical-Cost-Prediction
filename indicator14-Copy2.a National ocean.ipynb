```python
# Defining function for line plot
def line_plot(GeoAreaName):
    data = df8.loc[GeoAreaName]

    plt.plot(data.index, data.values)
```

```python
# a) Space fixed time varying
# PLot for Bulgaria country
line_plot('Norway')
plt.title("Line plot for country Norway of ocean's science expediture less")
plt.xlabel('Years')
plt.ylabel('Expenditure for ocean science per year')
```

```python
# b) bothe time and space varying
# PLots for top 5 countries in 5th target
plt.figure(figsize = (18, 8))

countries = ['Peru','Mozambique','Norway','Ireland','Oman']

for country in countries:
    line_plot(country)
    
plt.title("Line plot for five countries have less ocean's science expediture")
plt.xlabel('Years')
plt.ylabel('Expenditure for ocean science per year')
plt.legend(countries)
```

```python
Peru = indi_14_1_1.loc[indi_14_1_1['GeoAreaName']=='Peru']
Peru.head()
```

```python
#converting string date time into python date time object
Peru['TimePeriod'] = pd.to_datetime(Peru['TimePeriod'],format='%Y', errors='coerce')
```

```python
# Setting date as index
Peru_new=Peru.set_index('TimePeriod')
Peru_new.head()
```

```python
# Trend seasonal plot for Peru
from pylab import rcParams
import statsmodels.api as sm
plt.rcParams["figure.figsize"] = [14,6]
decomposition = sm.tsa.seasonal_decompose(Peru_new['Value'], model='Additive', freq =2)
fig = decomposition.plot()
plt.title('Seasonal plot for Peru')
plt.show()
```
