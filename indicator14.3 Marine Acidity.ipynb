```python
# Defining function for line plot
def line_plot(GeoAreaName):
    data = df8.loc[GeoAreaName]

    plt.plot(data.index, data.values)
```

```python
# a) Space fixed time varying
# PLot for Bulgaria country
plt.rcParams["figure.figsize"] = [10,6]
line_plot('Netherlands')
plt.title('Line plot for country Netherlands has the more ocean acidification ')
plt.xlabel('Years')
plt.ylabel('pH of marine')
```

```python
# b) bothe time and space varying
# PLots for top 5 countries which has high Ph
plt.figure(figsize = (12,8))

countries = ['Netherlands','Kenya','Philippines','Mexico','Italy']

for country in countries:
    line_plot(country)
    
plt.title('Line plot for five countries have the more ocean acidification ')
plt.xlabel('Years')
plt.ylabel('pH of marine')
plt.legend(countries)
```
