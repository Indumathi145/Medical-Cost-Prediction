```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
import matplotlib as mlt
```

```python
## Importing dataset
df = pd.read_excel("Goal_14.xlsx")
df.head()
```

```python
df.isnull().sum()
```

```python
# To know the empty columns
df.count()
```

```python
# Removing empty columns
df1= df.drop(['TimeCoverage','Type of support','Type of waste treatment', 'Type of skill', 'Substance use disorders', 'Report Ordinal', 'Parliamentary committees', 'Policy Domains', 'Policy instruments', 'Migratory status', 'Mode of transportation', 'Mountain Elevation', 'Name of international institution','Name of non-communicable disease','Grounds of discrimination','Hazard type','IHR Capacity','Level of requirement','Level/Status','Fiscal intervention stage','Food Waste Sector','Counterpart','Cause of death','GeoInfoUrl','BasePeriod','TimeCoverage'],axis=1)
df1.head()
```

```python
# Removing unsignificant columns
df2 = df1.drop(['Disability status','Activity','Education level','Quantile','Cities','Sex','Age','Type of product'], axis=1)
df2.head()
```

```python
# Set indicator as index
df3 = df2.set_index(['Indicator'])
df3.head()
```

```python
# Removing unsignicant indicators
df4= df3.drop(['14.4.1','14.6.1','14.7.1','14.b.1'])
df4.head()
```

```python
df4.isnull().sum()
```

```python
df5 = df4.fillna(0)
df5.head()
```

```python
df5.isnull().sum()
```

```python
df5.info()
```

```python
# Converting object to float
df5['Value'] = df5['Value'].astype(float)
df5['TimePeriod']= df5['TimePeriod'].astype(object)
```

```python
df5.info()
```

```python
# Summary
df5.describe()
```

```python
df5['SeriesDescription'].unique()
```

## India Analysis

```python
df_i= df5.loc[df5['GeoAreaName']=='India']
df_i.head()
```

```python
df_i= df5.loc[df5['GeoAreaName']=='India']
Ind_pivot14_1=pd.pivot_table(df_i,index=["SeriesDescription","TimePeriod"],values=["Value"])
cm = sns.light_palette("orange", as_cmap=True)
Ind_pivot14_1.style.background_gradient(cmap=cm)
```

```python
# Defining function for line plot
def ind_line_plot(SeriesDescription):
    data2 = Ind_pivot14_1.loc[SeriesDescription]
    
    plt.plot(data2.index, data2.values)
```

```python
# World
df_world= df5.loc[df5['GeoAreaName']=='World']
wor_pivot14_1=pd.pivot_table(df_world,index=["SeriesDescription","TimePeriod"],values=["Value"])
wor_pivot14_1
```

```python
# Defining function for line plot
def wor_line_plot(SeriesDescription):
    data = wor_pivot14_1.loc[SeriesDescription]
    
    plt.plot(data.index, data.values)
```

```python
ind_wor=pd.concat([df_world,df_i],axis=0)
ind_wor.head()
```

```python
Geo_value1 = ind_wor[["GeoAreaName","Value"]]
Geo_value1_mean = Geo_value1.groupby(Geo_value1["GeoAreaName"]).mean().sort_values(by='Value', ascending=False)
Geo_value1_mean
```

```python
ind_wor_pivot14_2=pd.pivot_table(df_i,index=["SeriesDescription","TimePeriod"],values=["Value"])
ind_wor_pivot14_2
```

```python
# Indicator 1
ind_wor14_1=ind_wor.loc[ind_wor['SeriesDescription']=='Average proportion of Marine Key Biodiversity Areas (KBAs) covered by protected areas (%)']
```

```python
ind_wor_pivot14_3=pd.pivot_table(ind_wor14_1,index=["GeoAreaName","TimePeriod"],values=["Value"])
ind_wor_pivot14_3
```

```python
# Defining function for line plot
def worind_TI_line_plot(SeriesDescription):
    data = ind_wor_pivot14_3.loc[SeriesDescription]
    
    plt.plot(data.index, data.values)
    
```

```python
worind_TI_line_plot("India")
plt.title("Line plot for Average proportion of Marine Key Biodiversity Areas of India")
plt.xlabel('Years')
plt.ylabel('Average proportion of Marine Key Biodiversity Areas')
```

```python
worind_TI_line_plot("World")
plt.title("Line plot for Average proportion of Marine Key Biodiversity Areas of World")
plt.xlabel('Years')
plt.ylabel('Average proportion of Marine Key Biodiversity Areas')
```

```python
plt.figure(figsize = (18, 8))

gn=["India","World"]
for countinent in gn:
    worind_TI_line_plot(countinent)
plt.title("Line plot for comparing Average proportion of Marine Key Biodiversity Areas of India andWorld level and India")
plt.xlabel('Years')
plt.ylabel('Average proportion of Marine Key Biodiversity Areas')

plt.legend(gn)
```

```python
## Indidual bar plot for India 
ii= df_i.loc[df_i['SeriesDescription']=='Average proportion of Marine Key Biodiversity Areas (KBAs) covered by protected areas (%)']
Geo_valueI = ii[["TimePeriod","Value"]]
Geo_value_meanI = Geo_valueI.groupby(Geo_valueI["TimePeriod"]).mean()
Geo_value_meanI
```

```python

plt.rcParams["figure.figsize"] = [14,6]
ax=Geo_value_meanI.plot(kind='bar',title="Average proportion of Marine Key Biodiversity Areas of India for each year",color='orange')
ax.set_xlabel("Years")
ax.set_ylabel("Average proportion of Marine Key Biodiversity Areas")
```

```python
## Indidual bar plot for World
w= df_world.loc[df_world['SeriesDescription']=='Average proportion of Marine Key Biodiversity Areas (KBAs) covered by protected areas (%)']
Geo_valueW = w[["TimePeriod","Value"]]
Geo_value_meanW = Geo_valueW.groupby(Geo_valueW["TimePeriod"]).mean()
plt.rcParams["figure.figsize"] = [14,6]
x=Geo_value_meanW.plot(kind='bar',title="Average proportion of Marine Key Biodiversity Areas of World for each year",color='violet')
x.set_xlabel("Years")
x.set_ylabel("Average proportion of Marine Key Biodiversity Areas")
```

## Time series Analysis:

```python
# Autocorrelation plot for world

from pandas.plotting import autocorrelation_plot
autocorrelation_plot(w['Value'])
plt.title('The autocorrelation plot for Average proportion of Marine Key Biodiversity Areas of World')
plt.show()
```

```python
#converting string date time into python date time object
w['TimePeriod'] = pd.to_datetime(w['TimePeriod'],format='%Y', errors='coerce')
```

```python
# Trend seasonal plot for Australia
from pylab import rcParams
import statsmodels.api as sm
plt.rcParams["figure.figsize"] = [14,6]
decomposition = sm.tsa.seasonal_decompose(w['Value'], model='Additive', freq =2)
fig = decomposition.plot()
plt.title('Seasonal plot for Average proportion of Marine Key Biodiversity Areas of world')
plt.show()
```
